```python
# Gravitational Constant
G = 6.67E-11
# Mass of the first
M1 = float( input("please Enter the Mass of the first \t"))
# Mass of the second
M2 = float( input("please Enter the Mass of the second \t"))
# the distance between them
R = float( input("please Enter the distance between \t"))
# Gravitational force
Fg = ((G * M1 * M2)/(R**2))
print ("Gravitational force is \t", Fg)
```
